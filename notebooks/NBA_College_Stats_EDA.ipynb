```python
numeric_cols = df.select_dtypes(include='number').columns

#Prints out the amount of outliers in the dataset
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(col, ": ", len(outliers), "outliers")
```

```python
#Comparing two columns' means
print(df['PTS'].mean())
print(df['G'].mean())
```

```python
#Comparing two columns' medians
print(df['PTS'].median())
print(df['G'].median())
```

```python
#Gets all central tendency stats at once
df.describe()
```

```python
#Creates folder for histogram plots
os.makedirs("reports/histogram_folder", exist_ok=True)
#Create histograms with mean and median lines
for col in numeric_cols:
    plt.figure(figsize=(8,4))
    plt.hist(df[col].dropna(), bins=40)
    plt.axvline(df[col].mean(), color='red', label='Mean')
    plt.axvline(df[col].median(), color='green', label='Median')
    plt.legend()
    plt.title(f"{col} Distribution")
    try:
        plt.savefig(f"reports/histogram_folder/{col}_mean_median_histogram.png", dpi=300)
    except FileNotFoundError as e:
        plt.savefig(f"reports/histogram_folder/WS40_mean_median_histogram.png", dpi=300)
    plt.close()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python
#Reads the .csv file
df = pd.read_csv(r"C:\Users\harri\OneDrive\Documents\DataScienceProjects\NBACareerPredictorML\data\EDA_Ready_NBAProspects.csv")

#Initial Inspection - Shape of the dataset
df.shape
```

```python
numeric_cols = df.select_dtypes(include='number').columns

#Prints out the amount of outliers in the dataset
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(col, ": ", len(outliers), "outliers")
```

```python
#Comparing two columns' means
print(df['PER'].mean())
print(df['3P%'].mean())
```

```python
#Comparing two columns' medians
print(df['PER'].median())
print(df['3P%'].median())
```

```python
#Gets all central tendency stats at once
df.describe()
```

```python
#Creates folder for histogram plots
os.makedirs("reports/prospect_histogram_folder", exist_ok=True)
#Create histograms with mean and median lines
for col in numeric_cols:
    plt.figure(figsize=(8,4))
    plt.hist(df[col].dropna(), bins=40)
    plt.axvline(df[col].mean(), color='red', label='Mean')
    plt.axvline(df[col].median(), color='green', label='Median')
    plt.legend()
    plt.title(f"{col} Distribution")
    try:
        plt.savefig(f"reports/prospect_histogram_folder/{col}_mean_median_histogram.png", dpi=300)
    except FileNotFoundError as e:
        plt.savefig(f"reports/prospect_histogram_folder/WS40_mean_median_histogram.png", dpi=300)
    plt.close()
```

```python
#Computes the Std, Var, Range, Skewness, and Kurtosis
for col in numeric_cols:
    print(f"{col}:")
    print(f"    std:    {df[col].std()}")
    print(f"    var:    {df[col].var()}")
    print(f"    range:    {df[col].max() - df[col].min()}")
    print(f"    skewness:   {df[col].skew()}")
    print(f"    kurtosis:    {df[col].kurt()}")
    print()
```

```python
#Creates a numeric only data frame
numeric_df = df.select_dtypes(include='number')
#Creates the correlation matrix
corr_matrix = numeric_df.corr()
strong_corr = corr_matrix[(corr_matrix > 0.6) | (corr_matrix < -0.6)]
#Creates the heat map plot
plt.figure(figsize=(12,10))
sns.heatmap(strong_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix for Numeric Columns")
plt.show()
plt.savefig("reports/heatmaps/college_correlation_matrix.png", dpi=300)
plt.close()
```

```python
#Univariate Analysis on Conference, Class, and Position - Histogram

#Creates the histogram for Class
plt.figure(figsize=(6,4))
plt.hist(df['Class'], bins=[-0.5, 0.5, 1.5, 2.5, 3.5], edgecolor='black')
plt.xticks([0,1,2,3], ['FR(18 & 19)', 'SO(20)', 'JR(21)', 'SR(22+)'])
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
plt.savefig("reports/prospect_histogram_folder/Class_distribution.png", dpi=300)
plt.close()
```

```python
#Creates the histogram for Position
plt.figure(figsize=(6,4))
plt.hist(df['Pos'], bins=[-0.5, 0.5, 1.5, 2.5], edgecolor='black')
plt.xticks([0,1,2], ['Guard', 'Forward', 'Center'])
plt.title("Position Distribution")
plt.xlabel("Position")
plt.ylabel("Count")
plt.show()
plt.savefig("reports/prospect_histogram_folder/Pos_distribution.png", dpi=300)
plt.close()
```

```python
#Creates the histogram for Conference
plt.figure(figsize=(6,4))
plt.hist(df['Conf'], bins=[-0.5, 0.5, 1.5, 2.5, 3.5], edgecolor='black')
plt.xticks([0,1,2,3], ['International', 'Other Colleges', 'G League', 'Power Five College'])
plt.title("Conference Distribution")
plt.xlabel("Conference")
plt.ylabel("Count")
plt.show()
plt.savefig("reports/prospect_histogram_folder/Conf_distribution.png", dpi=300)
plt.close()
```
